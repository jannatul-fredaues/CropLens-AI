```python
# Install required library
!pip install -q python-docx
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from docx import Document
from docx.shared import Inches

# ==========================================
# Paths & Parameters
# ==========================================
input_path = "/kaggle/input/datasets/jannatulferdaues/medicinal-flower-dataset/flower_dataset"
output_path = "/kaggle/working/MobileNetV2_output"
os.makedirs(output_path, exist_ok=True)

IMG_SIZE      = (224, 224)
BATCH_SIZE    = 32
EPOCHS        = 25
LEARNING_RATE = 0.000001

# ==========================================
# Separate generators for train / val
#   Training generator  → augmentation ON
#   Validation generator → augmentation OFF (only preprocessing)
#   This prevents data-leakage from augmented images reaching evaluation.
# ==========================================
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode="nearest"
)

# Validation must use its OWN datagen with NO augmentation
val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    input_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42
)

validation_generator = val_datagen.flow_from_directory(
    input_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False,   # IMPORTANT: keep False for correct confusion matrix
    seed=42
)

# ==========================================
# Class Weights
# ==========================================
classes      = list(train_generator.class_indices.keys())
labels       = train_generator.classes
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights_array))
print(f"Classes found: {classes}")
print(f"Class weights: {class_weights}")

# ==========================================
# Build Model
# ==========================================
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False   # Freeze all layers initially

x           = base_model.output
x           = GlobalAveragePooling2D()(x)
x           = Dense(1024, activation='relu')(x)
x           = Dropout(0.5)(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)
model       = Model(inputs=base_model.input, outputs=predictions)

model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

# ==========================================
# Callbacks
# ==========================================
best_model_path = os.path.join(output_path, 'best_model.keras')

checkpoint = ModelCheckpoint(
    best_model_path,
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=6,
    restore_best_weights=True,
    verbose=1
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-7,
    verbose=1
)

# ==========================================
# Phase 1 ─ Train with frozen base
# ==========================================
print("\n========== Phase 1: Training top layers (frozen base) ==========")
history1 = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    class_weight=class_weights,
    callbacks=[checkpoint, early_stop, reduce_lr],
    verbose=1
)


# ==========================================
# Merge both training histories for plotting
# ==========================================
def merge_histories(h1, h2):
    merged = {}
    for key in h1.history:
        merged[key] = h1.history[key] + h2.history.get(key, [])
    return merged

full_history = merge_histories(history1, history2)
total_epochs = len(full_history['accuracy'])
phase1_end   = len(history1.history['accuracy'])

# ==========================================
# Plot Accuracy & Loss (full training)
# ==========================================
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(full_history['accuracy'],     label='Train Accuracy',      marker='o', markersize=3)
plt.plot(full_history['val_accuracy'], label='Validation Accuracy', marker='o', markersize=3)
plt.axvline(x=phase1_end - 1, color='red', linestyle='--', label='Fine-tune start')
plt.title("Model Accuracy ")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(full_history['loss'],     label='Train Loss',      marker='o', markersize=3)
plt.plot(full_history['val_loss'], label='Validation Loss', marker='o', markersize=3)
plt.axvline(x=phase1_end - 1, color='red', linestyle='--', label='Fine-tune start')
plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
accuracy_loss_curve_path = os.path.join(output_path, 'accuracy_loss_curve.png')
plt.savefig(accuracy_loss_curve_path, dpi=150)
plt.show()
plt.close()
print(f"✅ Accuracy/Loss curve saved: {accuracy_loss_curve_path}")

# ==========================================
# Use load_model (not load_weights)
# ==========================================
print(f"\nLoading best model from: {best_model_path}")
best_model = load_model(best_model_path)

# ==========================================
# Evaluation
# ==========================================
print("\n========== Evaluating on Validation Set ==========")

validation_generator.reset()

eval_result = best_model.evaluate(validation_generator, verbose=1)
print(f"Validation Loss:     {eval_result[0]:.4f}")
print(f"Validation Accuracy: {eval_result[1]:.4f}")

validation_generator.reset()
Y_pred = best_model.predict(validation_generator, verbose=1)
y_pred = np.argmax(Y_pred, axis=1)
y_true = validation_generator.classes

report = classification_report(y_true, y_pred, target_names=classes)
print("\nClassification Report:\n", report)

# ==========================================
# Confusion Matrix
# ==========================================
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(max(10, len(classes)), max(8, len(classes) - 2)))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=classes,
    yticklabels=classes
)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix — MobileNetV2')
plt.tight_layout()
confusion_matrix_path = os.path.join(output_path, 'confusion_matrix.png')
plt.savefig(confusion_matrix_path, dpi=150)
plt.show()
plt.close()
print(f"✅ Confusion matrix saved: {confusion_matrix_path}")

# ==========================================
# Word Report
# ==========================================
doc = Document()
doc.add_heading('Classification Report — MobileNetV2 (CropLens)', 0)

doc.add_heading('Dataset & Training Config', level=1)
doc.add_paragraph(f"Dataset path:        {input_path}")
doc.add_paragraph(f"Image size:          {IMG_SIZE[0]} × {IMG_SIZE[1]}")
doc.add_paragraph(f"Batch size:          {BATCH_SIZE}")
doc.add_paragraph(f"Phase 1 epochs:      {phase1_end}")
doc.add_paragraph(f"Phase 2 epochs:      {total_epochs - phase1_end}")
doc.add_paragraph(f"Base learning rate:  {LEARNING_RATE}")
doc.add_paragraph(f"Fine-tune LR:        {LEARNING_RATE / 10}")
doc.add_paragraph(f"Classes:             {', '.join(classes)}")

doc.add_heading('Model Performance', level=1)
doc.add_paragraph(f"Best Validation Loss:     {eval_result[0]:.4f}")
doc.add_paragraph(f"Best Validation Accuracy: {eval_result[1] * 100:.2f}%")

doc.add_heading('Classification Metrics', level=1)
doc.add_paragraph(report)

doc.add_heading('Training History', level=1)
doc.add_paragraph(
    "Red dashed line marks the transition from Phase 1 (frozen base) to Phase 2 (fine-tuning)."
)
doc.add_picture(accuracy_loss_curve_path, width=Inches(6))

doc.add_heading('Confusion Matrix', level=1)
doc.add_picture(confusion_matrix_path, width=Inches(6))

report_path = os.path.join(output_path, 'classification_report_mobilenetv2.docx')
doc.save(report_path)
print(f"\n✅ MobileNetV2 training complete!")
print(f"📄 Report saved to: {report_path}")
```
